#### 27. Consider an integer vector Z, which of these expressions are legal? (★☆☆)
```python
Z**Z
2 << Z >> 2
Z <- Z
1j*Z
Z/1/1
Z<Z>Z
```

````python
import numpy as np
Z = np.array([3], dtype = np.int8)

print(Z ** Z)
print( 2 << Z >> 2) # x << y이면 x * 2**y이고, x >> y이면 x / 2**y이다.
print( Z<- Z) # Z < -Z라고 읽혀짐
print(1j * Z) 
print(Z/1/1)
print(Z<Z>Z)
````

#### 29. How to round away from zero a float array ? (★☆☆)

````python
Z = np.random.uniform(-10,+10,10)
print(Z) # 원래 값
print(np.copysign(np.ceil(np.abs(Z)), Z)) # 정답값

print(np.abs(Z)) # 절대값
print(np.ceil(np.abs(Z))) # 올림
# np.copysign => 해당 값을 기존의 Z의 부호대로 바꿔줌
````

#### 35. How to compute ((A+B)*(-A/2)) in place (without copy)? (★★☆)
(별도로 변수에 copy하지 않고 값을 저장할 수 있는가(A, B에))

````python
A = np.ones(3)*1
B = np.ones(3)*2
print(A, B)

np.add(A, B, out = B)
print(A, B)

np.divide(A, 2, out = A)
print(A, B)

np.negative(A, out = A)
print(A, B)

np.multiply(A, B, out = A)
print(A, B)
````

#### 38. Consider a generator function that generates 10 integers and use it to build an array (★☆☆)

````python
def generate():
    for x in range(10):
        yield x

a = np.fromiter(generate(), dtype = int)
print(a)
````
